```python
import pandas as pd
import numpy as np
# read the dataset from the link, extract and make a dataframe
import requests, zipfile, io
r = requests.get('https://archive.ics.uci.edu/static/public/438/health+news+in+twitter.zip', stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
p_data = open('Health-Tweets/bbchealth.txt')
myls=list(p_data)
mylist = [line.rstrip('\n') for line in myls]
print(mylist)
```

```python
import preprocessor as p
List=[]
for i in range(len(mylist)):
    text=p.clean(mylist[i])
    List.append(text)
print(List)
```

```python
tweet=[]
for d in List:
    p=d.split("|")
    tweet.append(p[2])

print(tweet)
```

```python
import sys
import json
import re
import copy
import math

regex_str = [
    r'<[^>]+>',
    r'(?:@[\w_]+)',
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)",
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+',
    r'(?:(?:\d+,?)+(?:\.?\d+)?)',
    r"(?:[a-z][a-z'\-_]+[a-z])",
    r'(?:[\w_]+)',
]

tokens_re = re.compile(r'(' + '|'.join(regex_str) + ')', re.VERBOSE | re.IGNORECASE)


def tokenize(s):
    return tokens_re.findall(s)


def preprocess(s, lowercase=True):
    tokens = tokenize(s)
    if lowercase:
        tokens = [token.lower() for token in tokens]
    return tokens
```

```python
tweet_all = []
for t in tweet:
    tokens = preprocess(t)
    tweet_all.append([term for term in tokens])
print(tweet_all)
l=len(tweet_all)

#preprocess(tweet[4])
```

```python
# defining the Jaccard distance
def jaccard(a, b):
    inter = list(set(a) & set(b))
    I = len(inter)
    union = list(set(a) | set(b))
    U = len(union)
    return round(1 - (float(I) / U), 4)


# updating the centroids 
def up_date(cluster, tweet_all, l, k):
    indices = []
    new_centxt_index = []
    new_centroid = []
    for i in range(k):
        indices.append([j for j, u in enumerate(cluster) if u == i])
        m = indices[i]
        # m gives the indices if the elements of every cluster k

        if (len(m) != 0):
            txt = [tweet_all[p] for p in m]
            sim = [[jaccard(txt[i], txt[j]) for j in range(len(m))] for i in range(len(m))]
            f1 = [sum(i) for i in sim]
        new_centxt_index.append(
            m[(f1.index(min([sum(i) for i in sim])))])  # index of the point closer to all the other points
    new_centroid = [tweet_all[x] for x in new_centxt_index]
    return new_centroid

# computing the sum of squared errors

def sse(cluster, centroids, tweet_all, k, l):
    indices1 = []
    cen_txt=centroids
    sum = 0
    for i in range(k):
        indices1.append([j for j, u in enumerate(cluster) if u == i])

        t = [tweet_all[x] for x in indices1[i]]
        for j in range(len(indices1[i])):
            sum = sum + math.pow(jaccard(t[j], cen_txt[i]), 2)
    print('The sum of squared error,SSE=', sum)

    

# output 

def output(cluster, k):
    final = []
    for i in range(k):
        final.append([j for j, u in enumerate(cluster) if u == i])
        t = [x for x in final[i]]
        print('Cluster ',i + 1,':', len(t),'tweets')   
    
# k-means implementation
def kmeans(centroids, tweet_all, l, k):
    count = 0
    for h in range(k):
        count = count + 1
        indices = [tweet_all.index(item) for item in centroids]
        cen_txt = [tweet_all[x] for x in indices]
        cluster = []
        for i in range(l):
            d = [jaccard(tweet_all[i], cen_txt[j]) for j in range(k)]
            ans = d.index(min(d))
            cluster.append(ans)

        centroids1 = up_date(cluster, tweet_all, l, k)
        sum = 0
        for i in range(k):
            if (centroids1[i] == centroids[i]):
                sum = sum + 1
        if (sum == k):
            break;
        centroids = copy.deepcopy(centroids1)

    output(cluster, k)
    sse(cluster, centroids, tweet_all, k, l)

```
